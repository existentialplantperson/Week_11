```python
#dependencies and setup
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import matplotlib.lines as mlines

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from scipy import signal
```

```python
df = pd.read_csv("multiTimeline.csv", parse_dates=True, index_col="Week")
df.shape
```

```python
df.head()
```

#### b. Decide whether ARMA, ARIMA, or SARIMA makes more sense for the dataset

I chose ARIMA for this data set. The data does not show seasonal change so SARIMA is not needed. The data is not stationary and so a model with order of integration is necessary. These two factors led me to ARIMA as the model for analysis.

```python
#data is not stationary or seasonal
plt.plot(df.index, df['Fortnite'])
```

```python
#autocorrelation used to estimate MA order (2)
plot_acf(df[['Fortnite']], lags=5, missing='drop')
```

```python
#PCAF used to estimate AR order (1)
plot_pacf(df[['Fortnite']], zero=False, lags = 10)
```

#### c. Run either an ARMA, ARIMA, or SARIMA on it

I ran an ARIMA model and used lag values of of AR = 1, I=1, and MA=1.

```python
#https://www.statsmodels.org/dev/generated/statsmodels.tsa.arima.model.ARIMA.html
#perform ARIMA because data is not stationary and is not seasonal
model = ARIMA(df['Fortnite'], order=(1, 1, 1))

#fit the model to the data
results = model.fit()

#create a dataframe
results_df = pd.DataFrame(results.forecast(steps=10)[0])
results_df.head(3)
```

```python
#ValueError: The computed initial AR coefficients are not stationary
#You should induce stationarity, choose a different model order, or you can
#pass your own start_params.
```

#### d. Explain the results

The plot of the ARIMA output shows a distinct decline in the trend of searches for Fortnite in the United States. The summary results show low values for the AIC, BIC, and P-values indicating that the model is a good fit for forcasting this data.

```python
results_df.plot(title="Fortnite searches in United States, 10 Day Forecast", 
                xlabel = 'Days', ylabel = 'Searches')
```

#### 2. What is a fourier transform? Perform one on y = np.cos(x/2), and both show (with code) and explain the key elements.

https://learn.adafruit.com/fft-fun-with-fourier-transforms/background

A fourier transform converts a data set from time series to frequency in order to understand the overall frequency trends and is used primarily for audio, time, and radar data.

The second graph below shows that the amplitude of the wave function is the most frequent intensity, dropping off sharply in either direction of the amplitude.

```python
x = np.arange(1,100,1)
y = np.cos(x/2)
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('Cosine Wave')
plt.show()
```

```python
psd = np.abs(np.fft.rfft(y))
freqs = np.fft.rfftfreq(len(y))
plt.plot(freqs, psd)
plt.axvline(1/(4*np.pi), c='r')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.title('Fast Fourier Transform on Cosine Wave')
handle = mlines.Line2D([], [], color='red', label='Amplitude')
plt.legend(handles=[handle])
plt.show()
```
